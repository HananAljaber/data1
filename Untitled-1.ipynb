```python
def pm(col):
    if col >50 :
        return "pass"
    else :return "faill"
df["passmath"]= df["math score"].apply(pm)
df["passwriting"]= df["writing score"].apply(lambda  x : "pass" if x > 50  else "faill")
df["passreading"]= df["reading score"].apply(lambda  x : "pass" if x > 50  else "faill")
df
```

```python
# df[["passmath","passwriting","passreading"]].
pm = df["passmath"].value_counts().reset_index()
pw = df["passwriting"].value_counts().reset_index()
pr = df["passreading"].value_counts().reset_index()
pm.columns = ["passmath", "count"]
sns.barplot(x="passmath",y="count",data=pm,palette=["red","blue"]);
```

```python
sns.barplot(x="passwriting",y="count",data=pw,palette=["#080","red"])
```

```python
sns.barplot(x="passreading",y="count",data=pr,palette=["#080","red"])
```

```python

numpath = df[df["math score"] == 100].groupby("gender")["math score"].count().reset_index()
numreading = df[df["reading score"] == 100].groupby("gender")["reading score"].count().reset_index()
numwriting = df[df["writing score"] == 100].groupby("gender")["writing score"].count().reset_index()
data = pd.DataFrame({
    "Subject": ["Math", "Reading", "Writing"],
    "Male": [
        numpath[numpath["gender"] == "male"]["math score"].values[0] if "male" in numpath["gender"].values else 0,
        numreading[numreading["gender"] == "male"]["reading score"].values[0] if "male" in numreading["gender"].values else 0,
        numwriting[numwriting["gender"] == "male"]["writing score"].values[0] if "male" in numwriting["gender"].values else 0
    ],
    "Female": [
        numpath[numpath["gender"] == "female"]["math score"].values[0] if "female" in numpath["gender"].values else 0,
        numreading[numreading["gender"] == "female"]["reading score"].values[0] if "female" in numreading["gender"].values else 0,
        numwriting[numwriting["gender"] == "female"]["writing score"].values[0] if "female" in numwriting["gender"].values else 0
    ]
})

data_long = data.melt(id_vars="Subject", var_name="Gender", value_name="Count")
data_long
# رسم المخطط
sns.barplot(x="Subject", y="Count", hue="Gender", data=data_long, palette="viridis")
# sns.barplot(x="Subject", y="Male", data=data, palette="viridis",hue="Female")

# # إعدادات الرسم
# plt.title("Number of Students Scoring 100 in Each Subject");
# plt.xlabel("Subject")
# plt.ylabel("Count")
```
